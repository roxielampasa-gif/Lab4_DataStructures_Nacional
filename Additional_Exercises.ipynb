```python
SEED_NUM = 3
STUDENT_MAJOR = "BSME" 
FAVORITE_ARTIST = "AESPA"
LOCAL_HAZARD = "CORRUPTION" 
CONTROL_NUM = max(3, SEED_NUM)

print(f"CONTROL_NUM: {CONTROL_NUM}")
```

Exercise 1: Unstructured Data Aggregator

```python
from collections import defaultdict, Counter

stream = []
for i in range(CONTROL_NUM + 3):
    stream.append(("TUP Manila", "BSCE" if i % 2 == 0 else "BSME"))


campus_map = defaultdict(list)
for campus, program in stream:
    campus_map[campus].append(program)

manila_counts = Counter(campus_map["TUP Manila"])
top_manila = manila_counts.most_common(1)[0]

print("--- EXERCISE 1 OUTPUT ---")
print(f"Expanded Stream Length: {len(stream)}")
print(f"Total Apps for TUP Manila: {len(campus_map['TUP Manila'])}")
print(f"Top Program at TUP Manila: {top_manila[0]}")
print(f"Frequency of Top Program: {top_manila[1]}")
```

Exercise 2: Jaccard Similarity Engine

```python
festival = {"Ben&Ben", "SB19", "Bini", "Eraserheads", FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"}
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

custom_artists = {f"Artist_Ref_{i}" for i in range(CONTROL_NUM)}
festival.update(custom_artists)

def get_jaccard(user_set, festival_set):
    intersection = user_set.intersection(festival_set)
    union = user_set.union(festival_set)
    return (len(intersection) / len(union)) * 100, len(intersection), len(union)

sim_a, inter_a, union_a = get_jaccard(user_a, festival)
sim_b, _, _ = get_jaccard(user_b, festival)

dealbreakers = user_a.difference(festival)

print("--- EXERCISE 2 OUTPUT ---")
print(f"Total Artists in Expanded Festival: {len(festival)}")
print(f"Length of Intersection (User A): {inter_a}")
print(f"Length of Union (User A): {union_a}")
print(f"User A Jaccard Similarity: {sim_a:.2f}%")
print(f"User B Jaccard Similarity: {sim_b:.2f}%")
print(f"User A Dealbreakers: {dealbreakers}")
```

Exercise 3: Stateful Sliding Window Filter

```python
from collections import deque, Counter

buffer = deque(maxlen=5)

burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD),
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

last_ts = burst[-1][0]
for i in range(1, CONTROL_NUM + 3):
    burst.append((last_ts + i, "TRAFFIC" if i % 2 == 0 else "FIRE"))

for alert in burst:
    current_categories = [a[1] for a in buffer]
    counts = Counter(current_categories)
    
    if counts[alert[1]] < 2:
        buffer.append(alert)

final_sum = sum(alert[0] for alert in buffer)

print("--- EXERCISE 3 OUTPUT ---")
print(f"Total Burst Size: {len(burst)}")
print(f"Final Deque Length: {len(buffer)}")
print(f"Final Sum of Timestamps: {final_sum}")
print(f"Exact Final Deque Output: {list(buffer)}")
```
